```python
import xgboost as xgb
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-2], df.iloc[:, -2:], test_size=0.2, random_state=42)
# Initialize the models
model1 = xgb.XGBClassifier(use_label_encoder=False,n_estimators=90,max_depth=3,learning_rate=0.1559,max_leaves=5, eval_metric='logloss')
model2 = xgb.XGBClassifier(use_label_encoder=False,n_estimators=170,max_depth=3,learning_rate=0.1559,max_leaves=5, eval_metric='logloss')

# Train the models
model1.fit(X_train, y_train['xyz_vaccine'])
model2.fit(X_train, y_train['seasonal_vaccine'])


```

```python
# Predict on the test set
y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)

# Combine the predictions into a DataFrame
predictions = pd.DataFrame({
    'xyz_vaccine': y_pred1,
    'seasonal_vaccine': y_pred2
})

# Evaluate the performance
accuracy1 = accuracy_score(y_test['xyz_vaccine'], y_pred1)
accuracy2 = accuracy_score(y_test['seasonal_vaccine'], y_pred2)
f1_1 = f1_score(y_test['xyz_vaccine'], y_pred1)
f1_2 = f1_score(y_test['seasonal_vaccine'], y_pred2)

print(f'Accuracy for xyz: {accuracy1:.4f}')
print(f'Accuracy for seasonal: {accuracy2:.4f}')
print(f'F1 Score for xyz: {f1_1:.4f}')
print(f'F1 Score for seasonal: {f1_2:.4f}')
```
